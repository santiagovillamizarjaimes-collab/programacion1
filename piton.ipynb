# Control de flujo en python

```python
# Series de Fibonacci:
# la suma de dos elementos define el siguiente
a, b = 0, 1
while b < 100:
  print(b)
  a, b = b, a+b # El valor de b no está actualizado (es importante escribirlo en comas)
```

```python
a, b = 0, 1
for b in range(20):
    print(b)
    a, b = b, a+b
```

El tipo más conocido de sentencia es el *if*

```python
x = int(input("Ingresa un número entero: "))
if x % 2 == 0:
  print("El número es par")
else:
  print("El número es impar.")
```

```python
x = int(input("Ingresa un entero, por favor: "))
if x < 0:
  x = 0
  print('Negativo cambiado a cero')
elif x == 0:
  print('Cero')
elif x == 1:
    print('Uno')
else:
  print('Mayor que uno')
```

```python
x = int(input("Ingrese un entero, por favor: "))
if x % 3 == 0:
   print("El número es múltiplo de 3")
else:
   print("El número NO es múltiplo de 3")
```

```python
x = int(input("Ingrese un entero, por favor: "))
if x % 3 == 0:
  print("El número es múltiplo de 3")
elif x % 3 == 1:
  print("El número no es múltiplo de 3 y tiene residuo 1")
elif x % 3 == 2:
  print("El número no es múltilo de 3 y tiene residuo 2")
```

En Python, break se usa para salir inmediatamente de un bucle (for o while) sin esperar a que termine normalmente.

```python
numeros = [3, 7, 12, 9, 5]
for n in numeros:
    if n == 12:
        print("¡Encontrado el número 12!")
        break
    print(n)
```
