```python
Clave_correcta = "python123"
intentos = 3

while intentos > 0:
  intento = input("Ingrese la contraseña: ")
  if intento == Clave_correcta:
    print("¡Acceso concedido!")
    break

  else:
    intentos -= 1
    print("Intentos restantes", intentos)

else:
  print("Cuenta bloqueada")
```

```python
Palabra = input("Ingrese una palabra:" )
vocales = "aeiou"
contador = 0

for letra in Palabra.lower():
  if letra in vocales:
    contador +=1

print("cantidad de vocales:", contador)
```

Pedir números hasta que el usuario ingrese un negativo.

 Solo sumar los números naturales

 Si ingresa texto, entonces ignorar (continue)

 Al final mostrar la suma


```python
suma = 0

while True:
    try:
        x = int(input("Ingrese un número entero: "))
    except ValueError:
        print("Entrada inválida. Ingrese solo números enteros.")
        continue
    if x < 0:
        break
    else:
        suma += x
print("La suma total es:", suma)
```
