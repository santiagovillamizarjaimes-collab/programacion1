# Estructuras en Python

```python
# Listas

lista = ["Manzana", "Platano", "Sandía", "Lista de frutas más consumidas"]

print(lista)
print(lista[-1])
print(lista[1:3])

lista[2] = "Naranja"

lista[2] = [3, 2, 1]
print("Lista modificada: ", lista)

print(len(lista))
```

```python
lista_nueva = [1, 2, 5, 7, 11]

print("Lista: ", lista_nueva)

lista_nueva.append(3)

print("Añadir el número primo faltante: ", lista_nueva)

print(lista_nueva.count(3))

print(lista_nueva.index(7))

lista_nueva.remove(3)

print("Lista original: ", lista_nueva)
```

```python
# Tuplas

tupla = ("¿La tierra es plana?", True, False)
print(tupla)

print(tupla[0])
print("Respuesta de un tierraplanista: ",tupla[1])
print("Repuesta de los científicos: ", tupla[2])

print(tupla.count(True))
print(tupla.index(False))
```

```python
# Conjunto

print(set())

print("Conjunto que contiene cinco multiplos de 5 =", set([5, 10, 15, 20, 25]))
print("Conjunto creado a partir de la tupla (5, 2, 5, 1, 1.5) = ", set((5, 2, 5, 1, 1.5)))
print("Conjunto creado a partir de la cadena '52511.5' = ", set("52511.5"))

conjunto = set([2, 3, 3, 4])
conjunto_2 = set([5, 3, 5, 6])
conjunto_3 = set([4, 2])

print("S= ", conjunto)
print("D= ", conjunto_2)
print("F=", conjunto_3)

conjunto.add(1)
print("Agregar el elemento 1 a S, ", conjunto)

conjunto.remove(1)
print("Imprimir el conjunto S original =", conjunto)

print("Intersección conjunto S y D", conjunto.intersection(conjunto_2))
```

```python
# Diccionario

diccionario = {1: "Uno", 2: "Dos"}
diccionario[3] = "Tres"
diccionario[4] = "Cuatro"
print(diccionario)

dict_lista_tuplas = dict([(1, "Uno"), (2, "Dos"), (3, "Tres")])
print("Tuplas escritas como un diccionario = ", dict_lista_tuplas)

dict_lista_string = dict(Uno = 1, Dos = 2, Tres = 3)
print(dict_lista_string)

dict_tipos = {1: "integer",
              2.2: "float",
              "texto": "string",
              (1, 2): "tupla"}

print("Diccionario de tipos de objetos en Python", dict_tipos)

dict_repeticion = {1: "Primero", 1: "Último"}
print(dict_repeticion)

print(diccionario,
      diccionario.keys(),
      diccionario.values(),
      diccionario.items())

claves = diccionario.values()
valores = diccionario.values()
print("Imprimir las claves de un diccionario= ", claves)
diccionario[1] = "One"
print("Imprimir los valores de un diccionario= ", valores)
diccionario.pop(2)
print(diccionario)
```

# Estructuras de control de flujo

```python
# Ejemplo de conversión de distintos tipos a False

print("¿0 es un objeto booleano?= ", bool(0))
print("¿0.0 es un objeto booleano?= ", bool(0.0))
print("¿"" es un objeto booleano?= ", bool(""))
print("¿Las listas son un objeto booleano?= ", bool(list()))
print("¿Las tuplas son un objeto booleano?= ", bool(tuple()))
print("¿Los conjuntos son un objeto booleano?= ", bool(set()))
print("¿Los diccionario son un objeto booleano?= ", bool(dict()))
print("¿None es un objeto booleano?= ", bool(None))
```

```python
# Ejemplo de conversión de distintos tipos a True

print(9000, "->", bool(9000))
print(3.1415, "->", bool(3.1415))
print("Cadena", "->", bool("Cadena"))
print(list((1, 2)), "->", bool(list((1, 2))))
print(tuple((1, 2)), "->", bool(tuple((1, 2))))
print(set((1, 2)), "->", bool(set((1, 2))))
print(dict([(1,2)]), "->", bool(dict([(1,1)])))
```

```python
# Ejemplo if-else

nota = float(input("Introduzca su nota: "))
if nota < 3.0:
  print("Estás reprobado, F")
else:
  print("Estás aprobado, ¡enhorabuena!")
```

```python
#Ejemplo if-elif-else

nota = float(input("Introduzca su nota: "))
if 0 <= nota and nota < 3:
  print("Estás reprobado, F")
elif 3 <= nota and nota < 3.8:
  print("Basico")
elif 3.8 <= nota and nota < 4.5:
  print("Notable")
elif 4.5 <= nota and nota <= 5:
  print("Sobresaliente")
else:
  print("Nota fuera de rango")
```

```python
# Ejemplo while

contador = 3
while contador != 0:
  contador = contador - 1
  print("Descomponer 4 en facores primos", contador)
```

```python
# Ejemplo básico del for

for iterador in ["Hola", "olá", "Hi"]:
  print(iterador)
```

```python
# for sobre estructuras de datos

lista = list((2, 1, 0))
tupla = tuple((2, 1, 0))
conjunto = set((2, 1, 0))
diccionario = dict(((2,"Dos"), (1, "Uno"), (0, "Cero")))

for iterador in diccionario:
  print(iterador)
```

```python
# Acceso por índices a una lista con un for

lista = ["resultado de 0/2", "resultado de 2/2", "resultado de 4/2",
         "resultado de 6/2", "resultado de 8/2", "resultado de 10/2",
         "resultado de 12/2", "resultado de 14/2"]

for indice in range(len(lista)):
  print(lista[indice], "=", indice)
```

```python
# for para visualizar el efecto de las sentencias break y continue

for numero in range(5):
  print(numero)
```

```python
# Visualización del efecto de la sentencia break

for numero in range(100):
  if numero ** 2 == 64:
    break
  print(numero)
```

```python
# Visualización del efecto de la sentencia continue

for numero in range(20):
  if numero % 2 == 0 :
    continue
  print(numero)
```

```python
# Uso del else en un while

intentos = 3
while intentos > 0:
  if input(">>> Escriba el número formado por los 4 primeros números primos:") == "2357":
    print("¡Correcta!")
    break
  intentos = intentos - 1
  print("Contraseña incorrecta")
else:
  print("Se te acabaron los intentos ;(")
```
