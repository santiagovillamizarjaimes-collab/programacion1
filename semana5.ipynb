```python
def sumar_matrices(A, B):  # Pseudo código
  filas = len(A)
  columnas = len(A[0])
  #Verificar que las dimensiones sean iguales
  if filas != len(B) or columnas != len(B[0]):
    return "ERROR: las matrices deben tener la misma dimensiones para sumarse."

  else:
    resultado = []
    for i in range(filas):
      fila = []
      for j in range(columnas):
        fila.append(A[i][j]+B[i][j])
      resultado.append(fila)
    return resultado
```

```python
A = [[1, 2], [3, 4]]
B = [[1, 3], [5, 6]]

sumar_matrices(A, B)
```

```python
def producto_Hadamard(A, B): #Pseudo código (producto de Hadamard)
  filas = len(A)
  columnas = len(A[0])
  # Verificar que las dimensiones sean iguales
  if filas != len(B) or columnas != len(B[0]):
    return "Error: Las matrices deben tener las mismas dimensiones para sumarse."

  else:
    resultado = []
    for i in range(filas):
      fila = []
      for j in range(columnas):
        fila.append(A[i][j] * B[i][j])
      resultado.append(fila)
  return resultado
```

```python
A = [[1, 2], [3, 4]]
B = [[1, 3], [5, 6]]

for i in producto_Hadamard(A, B):
  print(i)
```

```python
def sumar_matrices1(A, B):
    filas = len(A)
    columnas = len(A[0])

    if filas != len(B) or columnas != len(B[0]):
      return "Error: Las matrices deben tener las mismas dimensiones para sumarse."

    else:
      try:
        resultado = []
        for i in range(filas):
          fila = []
          for j in range(columnas):
            fila.append(float(A[i][j]) + float(B[i][j]))
        return resultado

      except ValueError:
        return "Debe ingresar solo números."
```

```python
A = [[4, 6], [5, 4]]
C = [[2, "a"], [5, 6]]

sumar_matrices1(A, C)
```
